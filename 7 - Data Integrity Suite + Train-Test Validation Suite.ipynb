```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("wenruliu/adult-income-dataset")

print("Path to dataset files:", path)
```

```python
ls $path
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(path + '/adult.csv')
#df.head()
```

```python
!pip install deepchecks -U --quiet
```

```python
from deepchecks.tabular.suites import data_integrity, train_test_validation
from deepchecks.tabular import Dataset
```

```python
ds = Dataset(df, label='income')  # target
#ds
```

```python
# Data Integrity Suite
integrity_suite = data_integrity()
integrity_results = integrity_suite.run(ds)
integrity_results.show()
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_ds = Dataset(train_df, label='income')
test_ds = Dataset(test_df, label='income')
```

```python
# Train-Test Validation Suite
validation_suite = train_test_validation()
validation_results = validation_suite.run(train_ds, test_ds)
validation_results.show()
```
