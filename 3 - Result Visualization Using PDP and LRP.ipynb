```python
data.drop(['Kitchen Qual','Exter Qual'],axis=1,inplace=True)
```

```python
k=['Overall Qual','Gr Liv Area']
target=['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x=data[k]
y=data[target]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(
    n_estimators=100,
    random_state=42)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
from sklearn.inspection import PartialDependenceDisplay
plt.figure(figsize=(8, 6))
PartialDependenceDisplay.from_estimator(regressor, x_train,k, grid_resolution=20)
plt.title("PDP for Feature: Median Income (MedInc)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
PartialDependenceDisplay.from_estimator(regressor, x_train, [("Overall Qual", "Gr Liv Area")], grid_resolution=20)
plt.title("2D PDP for Features:Overall Qual and ")
plt.show()
```
